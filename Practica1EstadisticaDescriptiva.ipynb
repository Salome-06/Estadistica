##d) Realice un filtrado de datos atípicos usando el criterio del rango intercuartílico (IQR).

```python
df_filtrado = df.copy()

# Calcular Q1, Q3 e IQR para ambas variables en un solo paso
Q1 = df_filtrado[['PESO (Gramos)', 'TALLA (Centímetros)']].quantile(0.25);
Q3 = df_filtrado[['PESO (Gramos)', 'TALLA (Centímetros)']].quantile(0.75);
IQR = Q3 - Q1;

# Definir los límites para considerar un dato como atípico
limite_inferior = Q1 - 1.5 * IQR;
limite_superior = Q3 + 1.5 * IQR;

# Filtrar el DataFrame para eliminar outliers en ambas variables
condicion = (
    (df_filtrado['TALLA (Centímetros)'] >= limite_inferior['TALLA (Centímetros)']) & (df_filtrado['TALLA (Centímetros)'] <= limite_superior['TALLA (Centímetros)']) &
    (df_filtrado['PESO (Gramos)'] >= limite_inferior['PESO (Gramos)']) & (df_filtrado['PESO (Gramos)'] <= limite_superior['PESO (Gramos)'])
);
df_sinatipicos = df_filtrado[condicion];

# Mostrar la cantidad de datos restantes
print(f"Número de datos restantes después de eliminar outliers: {len(df_sinatipicos)}")
```

## f) Realice un análisis de normalidad de las variables peso y talla después de eliminar los outliers, utilizando los tests de Shapiro-Wilk, Kolmogorov-Smirnov y Anderson-Darling. Incluya los gráficos QQ-plot y presente sus conclusiones.

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, kstest, anderson, jarque_bera

# Lista de variables a evaluar con los nombres correctos
columnas_fisicas = ['PESO (Gramos)', 'TALLA (Centímetros)']

# Diccionario para almacenar resultados
resultados_normalidad = {}

# Evaluar cada variable
for col in columnas_fisicas:
    # Usar df_sinatipicos ya que los outliers han sido eliminados
    datos = df_sinatipicos[col].dropna()

    # Shapiro-Wilk
    stat_sw, p_sw = shapiro(datos)

    # Kolmogorov-Smirnov
    media, desv = np.mean(datos), np.std(datos)
    # kstest requiere que le pasemos una distribución de referencia. Para normalidad, se usa 'norm'.
    # Los argumentos 'args' son para la media y desviación estándar de la distribución de referencia.
    stat_ks, p_ks = kstest(datos, "norm", args=(media, desv))

    # Anderson-Darling
    resultado_ad = anderson(datos, dist='norm')
    stat_ad = resultado_ad.statistic
    # Los valores críticos de Anderson-Darling son para diferentes niveles de significancia.
    # El tercer valor crítico (índice 2) corresponde al 5% de significancia por defecto.
    crit_ad_5 = resultado_ad.critical_values[2]
    # Si el estadístico es menor que el valor crítico, no se rechaza la hipótesis nula (normalidad).
    ad_normal = stat_ad < crit_ad_5

    # Jarque-Bera (requiere al menos 20 observaciones)
    if len(datos) >= 20:
        stat_jb, p_jb = jarque_bera(datos)
    else:
        stat_jb, p_jb = np.nan, np.nan # No se puede calcular

    # Guardar resultados
    resultados_normalidad[col] = {
        'Shapiro-Wilk': p_sw,
        'Kolmogorov-Smirnov': p_ks,
        'Anderson-Darling': f'Estadístico: {stat_ad:.4f}, Crítico (5%): {crit_ad_5:.4f}',
        'Jarque-Bera': p_jb,
        'Normal': sum([
            p_sw > 0.05 if not np.isnan(p_sw) else False,
            p_ks > 0.05 if not np.isnan(p_ks) else False,
            ad_normal,
            p_jb > 0.05 if not np.isnan(p_jb) else False
        ]) >= 3 # Considerar normal si al menos 3 de 4 tests (con p-value > 0.05 o AD_statistic < critical_value) lo indican
    }

# Mostrar resultados en tabla
print("Resultados de los tests de normalidad (valores p):\n")
for var, res in resultados_normalidad.items():
    print(f"Variable: {var.replace(' (Gramos)', '').replace(' (Centímetros)', '')}") # Eliminar unidades para el display
    print(f"  Shapiro-Wilk p = {res['Shapiro-Wilk']:.4f}")
    print(f"  Kolmogorov-Smirnov p = {res['Kolmogorov-Smirnov']:.4f}")
    print(f"  Anderson-Darling {res['Anderson-Darling']}")
    print(f"  Jarque-Bera p = {res['Jarque-Bera']:.4f}")
    if res['Normal']:
        print("  → Distribución normal (según mayoría de tests)\n")
    else:
        print("  → NO es distribución normal (según mayoría de tests)\n")
```
